# Crea un celda de código debajo y escribe allí tu programa

````python
# Inicializamos el contador de aciertos
puntos = 0

print("¡A jugar!")

# Pregunta 1: Número
respuesta1 = input("\n¿Cuántos mundiales tiene Brasil? ")
if respuesta1 == "5":
    print("¡Correcto!")
    puntos + 1
else:
    print("Incorrecto, son 7.")

# Pregunta 2: Texto
respuesta2 = input("\n¿Cuál es el primer nombre de Messi? ")
if respuesta2 == "Lionel":
    print("¡Correcto!")
    puntos += 1
else:
    print("No, es Potter.")

# Pregunta 3: Operación matemática
respuesta3 = input("\n¿Cuánto es 4*(5-4)? ")
if respuesta3 == "4":
    print("¡Correcto!")
    puntos += 1
else:
    print("No.")

# Pregunta 4: Selección múltiple
print("\n¿Quién marcó el gol que le dió a España el mundial'?")
print("1. David Villa\n2. Iniesta\n3. Xavi\n4. Ramos")
respuesta4 = input("? ")
if respuesta4 == "2":
    print("¡Correcto!")
    puntos += 1
else:
    print("No.")

# Pregunta 5: Selección múltiple
print("\n¿Cuántos equipos diferentes han conseguido ganar un campeonato de Liga de primera división?")
print("1. 20\n2. 9\n3. 15\n4. 4")
respuesta5 = input("? ")
if respuesta5 == "2":
    print("¡Correcto!")
    puntos += 1
else:
    print("No.")

# Cálculos finales
total_preguntas = 5
porcentaje = (puntos / total_preguntas) * 100

print("\n--- RESULTADOS ---")
print(f"Mis felicitaciones, ¡has conseguido {puntos} respuestas correctas!")
print(f"Eso es una puntuación del {porcentaje} por ciento.")
````
